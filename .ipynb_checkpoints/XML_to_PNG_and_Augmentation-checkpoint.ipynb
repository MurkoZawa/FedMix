```python
import os
import xml.etree.ElementTree as ET
from PIL import Image, ImageDraw

# Definisci il percorso della directory contenente i file XML di annotazione
annotation_directory = r'C:\Users\utente\Desktop\Università\Tesi magistrale\FedMix\data\Annotations'

# Definisci il percorso della directory di output per le immagini PNG
output_directory = r'C:\Users\utente\Desktop\Università\Tesi magistrale\FedMix\data\AnnotationsPNG'

# Assicurati che la directory di output esista, altrimenti creala
os.makedirs(output_directory, exist_ok=True)

# Scansiona tutti i file nella directory delle annotazioni
for filename in os.listdir(annotation_directory):
    if filename.endswith(".xml"):
        # Costruisci il percorso completo del file XML
        xml_filepath = os.path.join(annotation_directory, filename)
        
        # Parse il file XML di annotazione
        tree = ET.parse(xml_filepath)
        root = tree.getroot()
        
        # Estrai le informazioni di annotazione
        image_name = root.find('Annotation').get('Name')
        regions = root.find('Annotation').find('Regions')
        
        # Crea un'immagine vuota su cui disegnare le regioni annotate
        image_width = 1000  # Sostituisci con la larghezza effettiva dell'immagine
        image_height = 1000  # Sostituisci con l'altezza effettiva dell'immagine
        image = Image.new("RGB", (image_width, image_height), color="white")
        draw = ImageDraw.Draw(image)
        
        # Disegna le regioni annotate sull'immagine
        for region in regions.findall('Region'):
            vertices = [(float(vertex.get('X')), float(vertex.get('Y'))) for vertex in region.findall('Vertices/Vertex')]
            draw.polygon(vertices, outline="red", fill='red')
        
               # Estrai il nome del file XML
        xml_filename = os.path.splitext(filename)[0]
        
        # Salva l'immagine PNG nella directory di output con il nome del file XML
        output_filepath = os.path.join(output_directory, xml_filename + '.png')
        image.save(output_filepath)

```

```python
import os
from PIL import Image

# Definisci il percorso della directory contenente le immagini PNG
input_directory = r'C:\Users\utente\Desktop\Università\Tesi magistrale\FedMix\data\AnnotationsPNG'

# Definisci il percorso della directory di output per le immagini in bianco e nero
output_directory = r'C:\Users\utente\Desktop\Università\Tesi magistrale\FedMix\data\Annotations_BW'

# Assicurati che la directory di output esista, altrimenti creala
os.makedirs(output_directory, exist_ok=True)

# Scansiona tutti i file nella directory delle immagini PNG
for filename in os.listdir(input_directory):
    if filename.endswith(".png"):
        # Costruisci il percorso completo dell'immagine PNG
        input_filepath = os.path.join(input_directory, filename)
        
        # Carica l'immagine
        image = Image.open(input_filepath)
        
        # Converte l'immagine in scala di grigi
        image_gray = image.convert("L")
        
        # Inverti i colori (nero diventa bianco e viceversa)
        image_bw = image_gray.point(lambda x: 255 if x < 128 else 0, '1')
        
        # Salva l'immagine in bianco e nero nella directory di output
        output_filepath = os.path.join(output_directory, filename)
        image_bw.save(output_filepath)
```
